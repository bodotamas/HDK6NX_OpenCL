Poisson

```python
def random_poisson_1():
    template = r"""
    A Mikuláscsomagokban a szaloncukrok száma Poisson eloszlású {{k}} várható értékkel.
    \begin{itemize}
    \item Mennyi a valószínűsége, hogy 10-nél több szaloncukor lesz egy csomagban?
    \item 12 csomagot véletlenszerűen kiválasztva mennyi a valószínűsége, hogy mindegyikben lesz legalább 8 darab?
    \end{itemize}
    """
    ks = ['8', '9', '7']
    k = random.choice(ks)

    question = template 
    question = question.replace('{{k}}', k)

    return question
```

```python
def random_poisson_2():
    template = r"""
    Egy akciófilmben {{t}} percenként átlagosan {{n}} csúnya szó hangzik el.
    Feltételezzük, hogy a szavak előfordulása Poisson eloszlást követ.
    \begin{itemize}
    \item Mennyi a valószínűsége, hogy egy fél órás filmrészletben pontosan 10 csúnya szó hangzik el?
    \item Mennyi a valószínűsége, hogy 10 percen belül legalább 3 csúnya szó lesz a filmben?
    \end{itemize}
    """
    
    ts = ['10', '8', '12', '15']
    ns = ['3', '4', '5']
    
    t = random.choice(ts)
    n = random.choice(ns)
    
    question = template
    question = question.replace('{{t}}', t)
    question = question.replace('{{n}}', n)

    return question
```
