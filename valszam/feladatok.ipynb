## Klasszikus valószínűségi mező

```python
def random_classic_1():
    template = r"""
    Két dobókockával dobva mennyi a valószínűsége, hogy a dobások különbségének négyzete
    \begin{itemize}
    \item {{count_question}}?
    \item {{interval_question}}?
    \end{itemize}
    """
    count_questions = [
        'kisebb, mint 9',
        'kisebb, mint 14',
        'nagyobb, mint 9',
        'nem nagyobb, mint 9',
        'pontosan 2',
        'nem 4',
        'páros szám',
        'páratlan szám'
    ]
    interval_questions = [
        'legalább 5, de legfeljebb 9',
        'nagyobb, mint 4 de kevesebb, mint 16',
        'a $[10, 100]$ intervallumon van',
        'a $[-5, 5]$ intervallumon van',
        'a $[0, 8]$ intervallumon van',
        'a $[4, 16)$ intervallumon van',
        'a $(9, 25)$ intervallumon van',
        'a $(9, 25]$ intervallumon van',
        'a $[9, 25)$ intervallumon van',
        'a $[9, 25]$ intervallumon van'
    ]
    count_question = random.choice(count_questions)
    interval_question = random.choice(interval_questions)
    question = template
    question = question.replace('{{count_question}}', count_question)
    question = question.replace('{{interval_question}}', interval_question)
    return question
```

```python
def random_classic_2():
    template = r"""
    Egy autóversenyen 16-an indulnak. A kedvenc versenyzőnk biztosan benne lesz az első {{N}}.
    Feltételezve, hogy mást nem tudunk a verseny alakulásáról, mennyi a valószínűsége, hogy
    \begin{itemize}
    \item a versenyzőnk benne lesz az első {{first}}?
    \item a versenyzőnk pontosan a {{rank}} helyezést nyeri el?
    \end{itemize}
    """

    Ns = [
        '8-ban', '10-ben', '12-ben'
    ]
    firsts = [
        '3-ban', '4-ben', '6-ban'
    ]
    ranks = [
        '2.', '3.', '4.'
    ]
    N = random.choice(Ns)
    first = random.choice(firsts)
    rank = random.choice(ranks)
    question = template
    question = question.replace('{{N}}', N)
    question = question.replace('{{first}}', first)
    question = question.replace('{{rank}}', rank)
    return question
```

## Geometriai valószínűség

```python
def random_geometry_1():
    template = r"""
    Egy kikötőbe két hajó egymástól függetlenül érkezik {{interval}} óra között.
    Ha mindkettő az érkezésük után pontosan {{delay}} órát várakozik ott, akkor
    \begin{itemize}
    \item mennyi a valószínűsége, hogy az említett időintervallumon belül egyszerre lesznek majd a kikötőben?
    \item Mennyi a valószínűsége, hogy {{side}} egyik hajó sem lesz a kikötőben?
    \end{itemize}
    """
    intervals = [
        '10 és 14', '8 és 13', '8 és 16', '11 és 16'
    ]
    delays = [
        '1', '2', 'másfél'
    ]
    sides = [
        '12 előtt', '12 után'
    ]
    interval = random.choice(intervals)
    delay = random.choice(delays)
    side = random.choice(sides)
    question = template
    question = question.replace('{{delay}}', delay)
    question = question.replace('{{interval}}', interval)
    question = question.replace('{{side}}', side)
    return question
```

```python
def random_geometry_2():
    template = r"""
    A $[0, 1]$ intervallumról véletlenszerűen kiválasztunk egy $a$ és egy $b$ pontot.
    \begin{itemize}
    \item Mennyi a valószínűsége, hogy ${{inequality}}$?
    \item Készítsen ábrát a kedvező és az összes eset szemléltetésére!
    \end{itemize}
    """
    inequalities = [
        'a + b < 1',
        'a + b < 0.7',
        'a + b > 0.5',
        '|a - b| < 0.4',
        '|a - b| < 0.5',
        '|a - b| < 0.6',
        'a^2 > b',
        '\sqrt{a} < b',
        'a < sqrt{b}',
        'a^2 < b^2',
        '\sqrt{a} < \sqrt{b}',
        'a \cdot b < 0.5',
        'a \cdot b > 0.64'
    ]
    inequality = random.choice(inequalities)
    question = template
    question = question.replace('{{inequality}}', inequality)
    return question
```

## Feltételes valószínűség, átrendezés

```python
def random_conditional_1():
    template = r"""
    Ismerjük a $P(A) = {{p1}}, P(A|B) = {{p2}}, P(B|A) = {{p3}}$ valószínűségeket.
    \begin{itemize}
    \item Határozzuk meg a $P(A \cup B)$ valószínűséget!
    \item Mennyi lesz a $P(\overline{A}|\overline{B})$ értéke?
    \end{itemize}
    """

    params = [
        (r'0.3', r'\dfrac{1}{3}', r'\dfrac{2}{3}'),
        (r'\dfrac{7}{12}', r'\dfrac{5}{9}', r'\dfrac{5}{7}'),
        (r'\dfrac{13}{20}', r'\dfrac{3}{13}', r'\dfrac{3}{8}'),
        (r'\dfrac{9}{20}', r'\dfrac{2}{3}', r'\dfrac{4}{9}'),
        (r'\dfrac{1}{2}', r'\dfrac{5}{8}', r'\dfrac{5}{12}'),
        (r'\dfrac{9}{15}', r'\dfrac{6}{7}', r'\dfrac{2}{3}')
    ]

    p1, p2, p3 = random.choice(params)
    question = template
    question = question.replace('{{p1}}', p1)
    question = question.replace('{{p2}}', p2)
    question = question.replace('{{p3}}', p3)
    return question
```

## Feltételes valószínűség, Bayes tétel

```python
def random_bayes_1():
    template = r"""
    Az $n!$ nevű tehetségkutató verseny 4 fordulós.
    (A 4. fordulót döntőnek nevezzük.)
    Az egyes fordulókon a versenyzők {{percentage}} jut át.
    \begin{itemize}
    \item Mennyi a valószínűsége, hogy ha valaki átjutott {{alternative}}?
    \item Mennyi a valószínűsége, hogy valaki {{final}}?
    \end{itemize}
    """
    percentages = [
        r'80\%-a, 60\%-a, 20\%-a és 10\%-a',
        r'50\%-a, 40\%-a, 30\%-a és 10\%-a',
        r'60\%-a, 30\%-a, 20\%-a és 10\%-a',
        r'80\%-a, 40\%-a, 20\%-a és 10\%-a',
        r'40\%-a, 50\%-a, 20\%-a és 10\%-a',
        r'70\%-a, 30\%-a, 20\%-a és 10\%-a',
        r'60\%-a, 20\%-a, 20\%-a és 10\%-a'
    ]
    alternatives = [
        'az első fordulón, akkor a harmadikon is át fog',
        'az első fordulón, akkor a negyediken is át fog',
        'a második fordulón, akkor a harmadikon is át fog',
        'a második fordulón, akkor a negyediken is át fog'
    ]
    finals = [
        'kiesik a versenyből',
        'bejut a döntőbe',
        'nem jut be a döntőbe',
        'eljut a 3. fordulóig'
    ]
    percentage = random.choice(percentages)
    alternative = random.choice(alternatives)
    final = random.choice(finals)
    question = template
    question = question.replace('{{percentage}}', percentage)
    question = question.replace('{{alternative}}', alternative)
    question = question.replace('{{final}}', final)
    return question
```
